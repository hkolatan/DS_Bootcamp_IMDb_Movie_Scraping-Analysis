```python
import csv
import pandas as pd

df_1 = pd.read_csv("imdb_scraped_data_action.csv")
df_2 = pd.read_csv("imdb_scraped_data_comedy.csv")
df_3 = pd.read_csv("imdb_scraped_data_crime.csv")
df_4 = pd.read_csv("imdb_scraped_data_drama.csv")
df_5 = pd.read_csv("imdb_scraped_data_horror.csv")
df_6 = pd.read_csv("imdb_scraped_data_thriller.csv")
df_7 = pd.read_csv("imdb_scraped_data_scifi.csv")
df_combined = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7])
```

```python
df_combined
```

```python
df_combined = df_combined.drop(columns=["Number"])
df_combined
```

```python
df_combined["Year"] = df_combined["Year"].str.replace(r'\D', '', regex=True).astype(str)
df_combined["Year"] = df_combined["Year"].str.replace(r'', '')
df_combined["Year"] = df_combined["Year"].str.replace(r'I', '', regex=True)
df_combined
```
